```python
import pandas as pd, numpy as np
```

```python
air_quality_df= pd.read_csv("combined_air_quality_data.csv", parse_dates=["Date"])
air_quality_df.head()
```

**Adding Weather Data to the dataset**

```python
file_newyork= pd.read_csv('data/open-meteo-40.74N74.04W51m.csv', skiprows=3, delimiter=",",parse_dates=["time"])
file_chicago= pd.read_csv('data/open-meteo-41.86N87.65W179m.csv', skiprows=3, delimiter=",",parse_dates=["time"])
file_losangeles= pd.read_csv('data/open-meteo-34.06N118.24W91m.csv', skiprows=3, delimiter=",",parse_dates=["time"])
```

```python
file_newyork["County Name"]= "New York"
file_chicago["County Name"]= "Cook"
file_losangeles["County Name"]= "Los Angeles"

weather_df= pd.concat([file_newyork, file_chicago, file_losangeles], ignore_index=True)
weather_df.head()
```

```python
weather_df["Hour"]= weather_df["time"].dt.hour
weather_df["Date"]= pd.to_datetime(weather_df["time"].dt.date)
weather_df.drop(["time"], axis=1, inplace=True)
weather_df.head()
```

```python
final_df= pd.merge(air_quality_df, weather_df, on=["Date", "Hour", "County Name"])
final_df.head()
```

**Adding Traffic Data for New York City**

```python
traffic_df= pd.read_csv("data/MTA_Bridges_and_Tunnels_Hourly_Crossings__Beginning_2019.csv",parse_dates=["Transit Timestamp"])

traffic_df
```

```python
print(traffic_df['Facility'].unique())
```

```python
#Only selecting facilities associated with New York County
manhattan_df= traffic_df[((traffic_df["Facility"]=='Queens Midtown Tunnel')|(traffic_df["Facility"]=='Brooklyn Battery Tunnel')|
                   (traffic_df["Facility"]=='Henry Hudson Bridge')|(traffic_df["Facility"]=='Triborough Manhattan Bridge'))
                   &(traffic_df["Transit Timestamp"].dt.month==1)&(traffic_df["Transit Timestamp"].dt.year.between(2020,2024))]

manhattan_df
```

```python
df_avg= manhattan_df.groupby("Transit Timestamp", as_index=False)["Traffic Count"].mean()

df_avg["Date"]= pd.to_datetime(df_avg["Transit Timestamp"].dt.date)
df_avg["Hour"]= df_avg["Transit Timestamp"].dt.hour

df_avg= df_avg[["Date", "Hour", "Traffic Count"]]
df_avg
```

```python
final_df["Average Traffic Density"]= None

final_df.loc[final_df["County Name"]=="New York", "Average Traffic Density"]= final_df.loc[final_df["County Name"]=="New York"]\
    .merge(df_avg, on=["Date", "Hour"], how="left")["Traffic Count"].values

final_df["Average Traffic Density"]= final_df["Average Traffic Density"].astype(float)

final_df
```

**Interpolating Missing Values**

```python
final_df["PM2.5"].interpolate(method="linear", inplace=True)

mask= final_df["County Name"]!="New York"
final_df.loc[mask, "PM10"]= final_df.loc[mask, "PM10"].interpolate(method="linear")
```

```python
#Updating the dataset
final_df.to_csv("updated_air_quality_dataset.csv", index=False)
```
